```python
import requests
from bs4 import BeautifulSoup
R=requests.get("https://www.makaan.com/pune-property/katraj-flats-for-sale-52167")
x=R.content
```

```python
Soup=BeautifulSoup(x,'html.parser')
print Soup.prettify()
```

```python
all=Soup.find_all('li',{'class':'cardholder'})
len(all)
```

```python
a=all[0].find_all('span',{'class':'val'})
type(a)
```

```python
a[1].text.replace('u','')
```

```python
l=[]
for items in all:
    d={}
    d["Price_in_Lac"]= (items.find_all('span',{'class':'val'})[1]).text#price
    d["Block(BHK)"]= (items.find('strong').find('span').text)#bhk
    d["Society"]= (items.find_all('strong')[1]).text#area/society
    d["Total_Area"]= (items.find('td',{'class','size'}).text)#area
    d["Rate/sq ft"]= (items.find('td',{'class','lbl rate'}).text.replace("/ sq ft",""))#rate /sq ft
    try:
        d["How_old"]= (items.find('li',{'class','keypoint'}).text.replace(" year old","").replace(" years old","").replace("Bathrooms","").replace("Possession by Dec 2019","1").replace("Possession by Apr 2021","NAN").replace("Resale ","NAN"))#years old
    except:
        d["How_old"]= 'NAN'
        
    try:    
        d["bathroom"]=(items.find_all('li',{'class','keypoint'})[1].text.replace(" Bathrooms","").replace("Resale ","1"))#bathrooms
    except:
        d["bathroom"]= '1'
        
    try:    
        d["floor"]= (items.find_all('li',{'class','keypoint'})[2].find('span').text.replace("Resale ","NAN"))#floor
    except:
        d["floor"]='NAN'
    l.append(d)
    
```

```python
import pandas as pd
df=pd.DataFrame(l)
data=df.drop(19)
print data
```

```python
data.to_csv('output.csv')
```
