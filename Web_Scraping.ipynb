```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.census2011.co.in/census/state/maharashtra.html")
c=r.content
#type(c)
```

```python
soup=BeautifulSoup(c,"html.parser")
print(soup.prettify())
type (soup)
```

```python
all=soup.find_all("table",{"class","table table1 table-striped table-hover "})
#print all
print type(all)
```

```python
print all
```

```python
datav=all[1].find_all("td")
print datav[1].text
```

```python
l=[]
for k in range(8):
    b={}
    k=k*3
    b["Description"]= datav[k].text
    b["Population"]= datav[k+1].text
    b["Percentage"]= datav[k+2].text
    l.append(b)
```

```python
import pandas as pd
df=pd.DataFrame(l)
#data=df.drop(19)
print df
```

```python
df.to_csv('output.csv')
```
